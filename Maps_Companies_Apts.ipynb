```python
Grid_map = folium.Map(location=[30.28607940673828, -97.73833465576172],zoom_start=13)

##Create a FeatureGroup of Marker
Competitors= folium.FeatureGroup(name='Competitors', control=True)
Landmarks = folium.FeatureGroup(name='Landmarks/Nature', control=True)
Company = folium.FeatureGroup(name='Companies', control=True)
Metro = folium.FeatureGroup(name='Major Metro Stops', control=True)
Apartments = folium.FeatureGroup(name='Apartments', control=True)

#Create a marker_cluster object as a child of pointsgroup:
marker_cluster1 = plugins.MarkerCluster().add_to(Competitors)
marker_cluster2 = plugins.MarkerCluster().add_to(Landmarks)
marker_cluster3 = plugins.MarkerCluster().add_to(Company)
marker_cluster4 = plugins.MarkerCluster().add_to(Metro)
marker_cluster5 = plugins.MarkerCluster().add_to(Apartments)
#Create Marker and add to marker_cluster

folium.Marker(
    location=[yelp.iloc[0].latitude, yelp.iloc[0].longitude],
    popup="Nervous Charlie's",
    icon=folium.Icon(icon="home",color="black")
).add_to(Grid_map)

    
i=0
while i< len(yelp):
    if 'Bagel' in yelp.iloc[i]['name']: 
        folium.Marker(
            location=[yelp.iloc[i].latitude, yelp.iloc[i].longitude],
            popup=f"<i>{yelp.iloc[i]['name']}</i>",
            icon=folium.Icon(icon="cloud",color="red"),
        ).add_to(marker_cluster1)
    else:
        pass
    
        
    tooltip = f"<i>{yelp.iloc[i]['name']}</i>"


    i+=1

i=0
while i<len(landmarks):
    folium.Marker(
        location=[landmarks.iloc[i].latitude, landmarks.iloc[i].longitude],
        popup=f"<i>{landmarks.iloc[i]['name']}</i>",
        icon=folium.Icon(icon="glyphicon-tree-conifer",color="green"),
    ).add_to(marker_cluster2)
    i+=1

i=0
while i<len(company):
    folium.Marker(
        location=[company.iloc[i].latitude, company.iloc[i].longitude],
        popup=f"<i>{company.iloc[i]['name']}</i>",
        icon=folium.Icon(icon="glyphicon-flag",color="blue"),
    ).add_to(marker_cluster3)
    i+=1    

i=0
while i<len(metro):
    folium.Marker(
        location=[metro.iloc[i].latitude, metro.iloc[i].longitude],
        popup=f"<i>{metro.iloc[i]['name']}</i>",
        icon=folium.Icon(color="purple"),
    ).add_to(marker_cluster4)
    i+=1    

i=0
while i<len(apartments):
    folium.Marker(
        location=[apartments.iloc[i].latitude, apartments.iloc[i].longitude],
        popup=f"<i>{apartments.iloc[i]['name']}</i>",
        icon=folium.Icon(color="orange"),
    ).add_to(marker_cluster5)
    i+=1  
    
#Use pointsgroup as the child of Map
Grid_map.add_child(Competitors)
Grid_map.add_child(Company)
Grid_map.add_child(Landmarks)            
Grid_map.add_child(Metro) 
Grid_map.add_child(Apartments) 

#Open the LayerControl of the map
folium.LayerControl().add_to(Grid_map)

#black = current NC location
#green = landmarks / nature
#red = major competitors
#blue = companies 
#purple = metro stops
#orange = food truck parks


Grid_map
```
