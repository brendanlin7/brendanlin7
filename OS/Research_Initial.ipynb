```python

#cleaned up code
api_base = 'https://services.nvd.nist.gov/rest/json/cves/1.0?'

endpoint = '?'
severity = 'cvssV3Severity=CRITICAL'
keyword = '&keyword=iOS' #macOS iOS Android'
isExactMatch = '&isExactMatch=true'
resultsPerPage = '&resultsPerPage='
per = 1000

response = requests.get(api_base + severity + keyword + isExactMatch + resultsPerPage + str(per))
data = response.json()
print(data['totalResults'])

total_list = []

count = 0
while count < per:
    if count == data['totalResults']:
        break
    else:
        inner_list = []
        description = data['result']['CVE_Items'][count]['cve']['description']['description_data'][0]['value']
        pubDate = data['result']['CVE_Items'][count]['publishedDate']
        
        inner_list.append(description)
        inner_list.append(pubDate)
        total_list.append(inner_list)
    count+=1
    
total_list
```
